```python
import pandas as pd

# Load the dataset
file_path = r'C:\Users\Kasper\OneDrive - Aarhus universitet\Dokumenter\Kandidat\2. Semester\Data Science Project\Data science project part 2\Football_data.csv'
football_data = pd.read_csv(file_path)

# Display the first few rows of the dataset
print(football_data.head())

# Display basic information about the dataset
print(football_data.info())

# Check for missing values
print(football_data.isnull().sum())
```

```python
from sklearn.preprocessing import LabelEncoder
import numpy as np

label_encoder = LabelEncoder()

teams = pd.concat([football_data['Home'], football_data['Away']]).unique()
teams_sorted = np.sort(teams)

# Fit the encoder on the sorted team names
label_encoder.fit(teams_sorted)

# Transform the 'Home' and 'Away' columns
football_data['Home'] = label_encoder.transform(football_data['Home'])
football_data['Away'] = label_encoder.transform(football_data['Away'])

# Optionally, to check the results
print(football_data[['Home', 'Away']].head())
```

```python
mapping = {'A': -1, 'D': 0, 'H': 1}

# Apply the mapping to the 'FTR' column
football_data['FTR'] = football_data['FTR'].map(mapping)

football_data.head(5)
```

```python
football_data = football_data.drop(columns=['Date', 'Day', 'Time', 'Season', 'Wk'])
```

```python
from sklearn.feature_selection import VarianceThreshold
#Define the threshold for variance
threshold = 0.01  # You can adjust this threshold based on your preference

#Initialize the VarianceThreshold object
selector = VarianceThreshold(threshold)

#Fit the selector to your DataFrame
selector.fit(football_data)

#Get the indices of the features to keep
indices_to_keep = selector.get_support(indices=True)

#Drop the variables with low variance
football_data = football_data.iloc[:, indices_to_keep]
```
